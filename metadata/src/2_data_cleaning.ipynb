```python
# Combine the two header rows into one clear name
weather.columns = [
    f"{top.strip()}_{bot.strip()}".replace(" ", "_").replace(".", "").replace(",", "")
    for top, bot in weather.columns
]

# Manually clean a few obvious ones
rename = {
    "Date_Date": "date",
    "Time_Time": "time",
    "Out_Temp": "temp_out",
    "Hi_Temp": "temp_high",
    "Low_Temp": "temp_low",
    "Out_Hum": "humidity_out",
    "Dew_Pt": "dew_point",
    "Wind_Speed": "wind_speed",
    "Wind_Dir": "wind_dir",
    "Rain_Rain": "rain_total",
    "Rain_Rate": "rain_rate",
    "Solar_Rad": "solar_rad",
    "Solar_Energy": "solar_energy",
}
weather = weather.rename(columns=rename)
weather
```

```python
rename = {
    "Unnamed:_0_level_0_Date": "date",
    "Unnamed:_1_level_0_Time": "time",
}
weather = weather.rename(columns=rename)
weather
```

```python
df_hourly = weather[weather['time'].map(lambda t: t.minute) != 30].copy()
df_hourly
```

```python
df_hourly.to_csv('FreemanRanchWeather2015_hourly_cleaned.csv')
```

```python
mapping = pd.read_excel('../TXSTDSC Photo & Donor Data for Mapping.xlsx').drop(['??'], axis=1)
mapping['PhotoDate'] = pd.to_datetime(
    mapping['PhotoYear'].astype(str) + '-' +
    mapping['PhotoMonth'].astype(str) + '-' +
    mapping['PhotoDay'].astype(str),
    errors='coerce'
)
mapping.drop(columns=['PhotoYear', 'PhotoMonth', 'PhotoDay'], inplace=True)
mapping
```

```python
df_merged = df_hourly.merge(mapping, right_on='PhotoDate', left_on='date', how='inner').drop_duplicates()
df_merged
```

```python
df_complete = df_merged.merge(df_2015, on='TXSTDonorNumber', how='inner').drop_duplicates()
df_complete
```

```python
# List the first three columns you want
first_cols = ['TXSTDonorNumber', 'date', 'time']

# Reorder the dataframe
df_complete = df_complete[first_cols + [col for col in df_complete.columns if col not in first_cols]]
df_complete
```

```python
df_complete = df_merged.merge(df_2015, on='TXSTDonorNumber', how='inner').drop_duplicates()
df_complete
```

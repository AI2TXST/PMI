```python
import pandas as pd

df_train = pd.read_csv("metadata_split/train_metaData.csv")
df_val = pd.read_csv("metadata_split/val_metaData.csv")
df_test = pd.read_csv("metadata_split/test_metaData.csv")

drop_cols = ["image", "filename", "stem", "Date", "donor_id", "unique_donor_id", "year.1"]

drop_cols_all = drop_cols

df_reduced_train = df_train.drop(columns=drop_cols_all, errors="ignore")
df_reduced_val = df_val.drop(columns=drop_cols_all, errors="ignore")
df_reduced_test = df_test.drop(columns=drop_cols_all, errors="ignore")

# print("Remaining columns:", df_reduced_train.columns.tolist())
print("Num remaining:", len(df_reduced_train.columns))
```

```python
import pandas as pd

df = df_reduced_train.copy()


label_map = {
    "FRESH": 0,
    "BLOAT": 1,
    "ACTIVE_DECAY": 2,
    "ADVANCED_DECAY": 3
}
df["class_label_encoded"] = df["class_label"].map(label_map)

# check mapping
print("Encoded target preview")
print(df[["class_label", "class_label_encoded"]].drop_duplicates())


numeric_cols = df.select_dtypes(include=["number"]).columns.tolist()
string_cols = df.select_dtypes(include=["object", "string"]).columns.tolist()


# print("\nNumeric columns")
# print(numeric_cols)

print("\nString columns")
print(string_cols)


print("\nString columns with unique counts")
for col in string_cols:
    print(f"{col}: unique={df[col].nunique(dropna=True)}, avg_len={df[col].astype(str).str.len().mean():.2f}")

target_col = "class_label_encoded"
raw_target_col = "class_label"


# input categorical columns only
categorical_cols = ["wind_dir", "season"]
categorical_cols = [c for c in categorical_cols if c in df.columns]

print("\nTarget column")
print(target_col)

print("\nInput categorical columns")
print(categorical_cols)


df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=False)
df_encoded = df_encoded.drop(columns=[raw_target_col], errors="ignore")

#uncomment to check
# print("\nFinal columns after one-hot encoding")
# print(df_encoded.columns.tolist())

print("\nFinal shape")
print(df_encoded.shape)

df_encoded.to_csv("./metadata_split/train_processed_data.csv", index=False)
print("\nSaved to train processed_data.csv")
```

```python
import numpy as np
import pandas as pd

train_df = pd.read_csv("metadata_split/train_processed_data.csv")
val_df   = pd.read_csv("metadata_split/val_processed_data.csv")
test_df  = pd.read_csv("metadata_split/test_processed_data.csv")

# identify numeric columns
numeric_cols = train_df.select_dtypes(include=[np.number]).columns.tolist()

# fill numeric NaNs using TRAIN medians
train_medians = train_df[numeric_cols].median()

train_df[numeric_cols] = train_df[numeric_cols].fillna(train_medians)
val_df[numeric_cols]   = val_df[numeric_cols].fillna(train_medians)
test_df[numeric_cols]  = test_df[numeric_cols].fillna(train_medians)

print("train total nan:", train_df.isna().sum().sum())
print("val total nan:", val_df.isna().sum().sum())
print("test total nan:", test_df.isna().sum().sum())

# # save cleaned versions
# train_df.to_csv("metadata_split/train_process_clean.csv", index=False)
# val_df.to_csv("metadata_split/val_process_clean.csv", index=False)
# test_df.to_csv("metadata_split/test_process_clean.csv", index=False)

# print("Saved cleaned CSVs.")
```

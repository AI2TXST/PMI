```python
import json


# dataset paths
TRAIN_JSON = "caption_splits/train_captions.json"
VAL_JSON   = "caption_splits/val_captions.json"
TEST_JSON  = "caption_splits/test_captions.json"


# Load function
def load_captions(json_path):
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    if isinstance(data, dict) and "data" in data:
        data = data["data"]

    return data  # list of dictionaries


# Load splits
train_data = load_captions(TRAIN_JSON)
val_data   = load_captions(VAL_JSON)
test_data  = load_captions(TEST_JSON)

print("Train samples:", len(train_data))
print("Val samples  :", len(val_data))
print("Test samples :", len(test_data))



# Inspect few examples
def show_examples(data, split_name, num_examples=1):
    print(f"\n {split_name} examples")
    for i, item in enumerate(data[:num_examples]):
        print(f"\nSample {i+1}")
        print("Label  :", item["label"])
        print("Caption:", item["caption"])


show_examples(train_data, "Train", num_examples=1)
show_examples(val_data, "Val", num_examples=1)
show_examples(test_data, "Test", num_examples=1)


# generator for training
def caption_generator(data):
    for item in data:
        label = item["label"]
        caption = item["caption"]
        yield caption, label


# label counts (uncomment to check)
# def count_labels(data):
#     counts = {}
#     for item in data:
#         label = item["label"]
#         counts[label] = counts.get(label, 0) + 1
#     return counts


# print("\nTrain label counts:", count_labels(train_data))
# print("Val label counts  :", count_labels(val_data))
# print("Test label counts :", count_labels(test_data))
```
